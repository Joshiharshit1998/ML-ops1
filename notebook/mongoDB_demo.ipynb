```python
client = pymongo.MongoClient(CONNECTION_URL)
data_base = client[DB_NAME]
collection = data_base[COLLECTION_NAME]
```

```python
# Uploading data to MongoDB
from pymongo.errors import OperationFailure

try:
    if not data:
        print("No records to insert")
    else:
        rec = collection.insert_many(data, ordered=False)
        print(f"Inserted {len(rec.inserted_ids)} documents")
except OperationFailure as err:
    print("MongoDB operation failed:", err)
    print("Check your CONNECTION_URL credentials, IP whitelist, and cluster permissions.")
except Exception as err:
    print("Unexpected error during insert_many:", err)
```
